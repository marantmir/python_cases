## Limpando os Dados

Olhando a planilha original, podemos ver que tem algumas colunas que podem ser removidas, pois os dados já estão categorizados em seu cabeçalho.

Veja:

<center><img width="100%" src="https://github.com/jovianosilveira/python_cases/blob/master/20200930_QueimadasBrasil/img/01.png?raw=true"></center>

Sabendo que temos essas colunas, podemos deleta-las ou selecionar somente as que queremos ver.
Eu recomendo que você use a prática de selecionar somente o que você quer ver.

Isso porque, suponhamos que a fonte de dados aumente, nesse caso, irão "surgir" colunas indesejadas.
Quando tivermos os dados de 2020, isso vai acontecer, por exemplo...

```python
colunas = ['Código CNUC', 
           'Nome da UC', 
           'Categoria da UC: sigla federal',
           'Categoria da UC: nomenclatura nacional', 
           'Grupo de Proteção',
           'Ano de criação', 
           'Coordenação Regional do ICMBio',
           'Área estimada da UC (ha)', 
           'Bioma referencial',
           'Área queimada em 2019', 
           'Área queimada em 2018',
           'Área queimada em 2017', 
           'Área queimada em 2016', 
           'Área queimada em 2015', 
           'Área queimada em 2014', 
           'Área queimada em 2013', 
           'Área queimada em 2012']
```

Se eu te falei anteriormente para capturar somente os dados que você vai usar, então, vamos carregar denovo a variavel "dados" somente com as colunas que queremos?

Boralá!

```python
dados = pd.read_excel(link_dataset, sheet_name = 'queima', usecols = colunas)
dados.head() # visualizar somente 5 primeiros
```

para que possamos analisar melhor, precisamos transformar as colunas "Área queimada em 20XX" em duas colunas, sendo uma para o ano, e outra para o valor, deixando assim o modelo mais parecido ao famoso modelo de BI.

Primeiramente, vamos alterar o nome das colunas, e para isso vamos usar um dicionário, para fazer uma espécie de DE: PARA:

```python
nomes_novos = {
    'Área queimada em 2019' : '2019', 
    'Área queimada em 2018' : '2018',
    'Área queimada em 2017' : '2017', 
    'Área queimada em 2016' : '2016', 
    'Área queimada em 2015' : '2015', 
    'Área queimada em 2014' : '2014', 
    'Área queimada em 2013' : '2013', 
    'Área queimada em 2012' : '2012'
}
```

```python
dados = dados.rename(columns = nomes_novos)
dados
```

```python
colunas_estaticas = ['Código CNUC', 
           'Nome da UC', 
           'Categoria da UC: sigla federal',
           'Categoria da UC: nomenclatura nacional', 
           'Grupo de Proteção',
           'Ano de criação', 
           'Coordenação Regional do ICMBio',
           'Área estimada da UC (ha)', 
           'Bioma referencial']

dados = dados.melt(id_vars= colunas_estaticas , var_name='Ano', value_name='Area')
dados
```

Você vai perceber que após fazer isso, os anos que não tinham queimada, aparecem como NaN (not a number = vazio).

Nosso trabalho agora é limpar isso.

```python
dados = dados.dropna(subset=['Area'])
dados
```

Vamos agora separar a sigla do estado (UF) do nome da Coordenação Regional do ICMBio, porém, antes de mais nada, vamos reiniciar o indice para evitar possiveis problemas.

```python
dados = dados.reset_index(drop=True) # com isso apaga o INDEX anterior, e substitui pelo novo
dados.head()
```

Agora sim vamos quebrar a coluna.
Vamos iniciar usando um método que se aplica a Serie (coluna) do pandas sem necessidade de aplicação de LAMBDA.

```python
dividir_coluna = dados['Coordenação Regional do ICMBio'].str.rsplit("/", expand=True)
dividir_coluna
```

vimos que as novas colunas geradas têm o nome 0 e 1, e precisamos substituir por um valor que faça sentido para nós.

E já aprendemos nesse artigo como fazer...Boralá denovo!

```python
dividir_coluna = dividir_coluna.rename(columns = {0:"Coordenação", 1: "UF"})
dividir_coluna.head()
```

como podemos ver, o indice é o mesmo do dataframe de onde pegamos a coluna para quebrar em duas.

Como fazemos para juntar isso? 

Não é tão fácil como beber água, mas quase...

```python
dados = dados.join(dividir_coluna)
dados.head()
```

Agora sim faz sentido deletarmos a coluna que não vamos mais usar, pois anteriormente a quebramos em duas

```python
dados = dados.drop(columns=['Coordenação Regional do ICMBio'])
dados.head()
```
